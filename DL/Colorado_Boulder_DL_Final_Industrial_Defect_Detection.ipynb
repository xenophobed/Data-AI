## Data Augmentation
<div class="alert alert-success" role="alert" style="border-radius:10px">
    <p>When training with image data without data augmentation, we simply need the specified number of data and create a mini-batch. When executing data augmentation, after acquiring the data, various augmentation techniques are applied to the image to create a new mini-batch. <br/>
    The main parameters of data augmentation techniques are as follows : </p>
    <ul>
        <li><b>rotation_range</b> : Rotate the image (ex. 50 -> rotate randomly in -50~50)</li>
        <li><b>zoom_range</b> : Zoom in/out on the image (ex. 0.5 -> zoom in/out randomly in 1-0.5~1+0.5)</li>
        <li><b>brightness_range</b> : Change the brightness (ex. [0.3,1.0] -> change randomly in [0.3,1.0])</li>
        <li><b>vertical_flip</b> : Flip the image upside down</li>
        <li><b>horizontal_flip</b> : Flip the image left or right</li>
        <li><b>height_shift_range</b> : Move the image up or down in parallel (ex. 0.3 -> move up/down randomly in [-0.3*Height, 0.3*Height])</li>
        <li><b>width_shift_range</b> : Move the image left or right in parallel (ex. 0.3 -> move left/right randomly in [-0.3*Width, 0.3*Width])</li>
        <li><b>rescale</b> : The image is normalized by multiplying each pixel value by a constant. (ex. 1/255 -> normalize the RGB value of each pixel between 0.0 and 1.0)</li>
    </ul>
</div>

```python
img = cv2.imread(train_path + 'ok_front/cast_ok_0_1016.jpeg')
img_4d = img[np.newaxis]
plt.figure(figsize=(25,10))
generators = {"rotation":ImageDataGenerator(rotation_range=180),
              "zoom":ImageDataGenerator(zoom_range=0.7),
              "brightness":ImageDataGenerator(brightness_range=[0.2,1.0]),
              "height_shift":ImageDataGenerator(height_shift_range=0.7),
              "width_shift":ImageDataGenerator(width_shift_range=0.7)}

plt.subplot(1, 6, 1)
plt.title("Original", weight='bold', size=15)
plt.imshow(img)
plt.axis('off')
cnt = 2
for param, generator in generators.items():
    image_gen = generator
    gen = image_gen.flow(img_4d, batch_size=1)
    batches = next(gen)
    g_img = batches[0].astype(np.uint8)
    plt.subplot(1, 6, cnt)
    plt.title(param, weight='bold', size=15)
    plt.imshow(g_img)
    plt.axis('off')
    cnt += 1
plt.show()
```

## Setting up the Image Data



```python
image_gen = ImageDataGenerator(rescale=1/255,
                               zoom_range=0.1,
                               brightness_range=[0.9,1.0])
```

- Directory: Images are loaded from train_path.
- Target Size: Images are resized to 300x300 pixels.
- Color Mode: Images are converted to grayscale.
- Classes: Defines class labels, where 'def_front' is labeled as 0 and 'ok_front' as 1.
- Class Mode: Set to 'binary' for binary classification.
- Shuffling: Images are shuffled to ensure model receives a random order during training.

```python
image_shape = (300,300,1) # 300  300graysclaed (full-color : 3)
batch_size = 32

train_set = image_gen.flow_from_directory(train_path,
                                            target_size=image_shape[:2],
                                            color_mode="grayscale",
                                            classes={'def_front': 0, 'ok_front': 1},
                                            batch_size=batch_size,
                                            class_mode='binary',
                                            shuffle=True,
                                            seed=0)

test_set = image_gen.flow_from_directory(test_path,
                                           target_size=image_shape[:2],
                                           color_mode="grayscale",
                                           classes={'def_front': 0, 'ok_front': 1},
                                           batch_size=batch_size,
                                           class_mode='binary',
                                           shuffle=False,
                                           seed=0)
```

## CNN Moel 4 binary classification

### Model Architecture
The model is defined sequentially with the following layers:

- Initial Convolutional Layer:

Filters: 16
Kernel Size: 7x7
Strides: 2
Activation: ReLU
Padding: 'same'
This layer is designed to extract features from the input image using 16 filters, with a stride of 2 for downscaling, and applies the ReLU activation function.
- First Max Pooling Layer:

Pool Size: 2x2
Strides: 2
Reduces the spatial dimensions of the output from the previous layer, helping to make the model more robust to variations in the input data.
- Second Convolutional Layer:

Filters: 32
Kernel Size: 3x3
Activation: ReLU
Padding: 'same'
Increases the depth of the network by capturing more complex features with 32 filters.
- Second Max Pooling Layer:

Same configuration as the first pooling layer, further reducing the spatial dimensions.
- Third Convolutional Layer:

Filters: 64
Kernel Size: 3x3
Activation: ReLU
Padding: 'same'
Continues to increase the depth, allowing the network to capture even more complex patterns.
- Third Max Pooling Layer:

Further reduces feature dimensions and prepares the data for the fully connected layers.
- Flatten Layer:

Converts the 3D feature maps into 1D feature vectors, making them suitable for processing in fully connected layers.
- Fully Connected Layer:

Units: 224
Activation: ReLU
Serves as a classification layer with ReLU activation.
- Dropout Layer:

Rate: 0.2
Helps prevent overfitting by randomly setting a fraction of input units to 0 during training.
- Output Layer:

Units: 1
- Activation: Sigmoid
Produces the final classification output. The sigmoid activation function outputs a value between 0 and 1, representing the probability of the input belonging to the positive class.

### Compilation
The model is compiled with the following configuration:

Loss Function: Binary Crossentropy, suitable for binary classification tasks.
Optimizer: Adam, a popular choice for training deep learning models due to its adaptive learning rate properties.
Metrics: Accuracy, to monitor the model's performance during training.
python
C

```python
backend.clear_session()
model = Sequential()
model.add(Conv2D(filters=16, kernel_size=(7,7), strides=2, input_shape=image_shape, activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
model.add(Conv2D(filters=32, kernel_size=(3,3), strides=1, input_shape=image_shape, activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
model.add(Conv2D(filters=64, kernel_size=(3,3), strides=1, input_shape=image_shape, activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
model.add(Flatten())
model.add(Dense(units=224, activation='relu'))
model.add(Dropout(rate=0.2))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.summary()
```

```python
plot_model(model, show_shapes=True, expand_nested=True, dpi=60)
```

## Build Model

###Setting Up Callbacks for Training
- Early Stopping

early_stop = EarlyStopping(monitor='val_loss', patience=2): This callback monitors the validation loss (val_loss) and will stop the training process if the validation loss doesn't improve for 2 consecutive epochs (patience=2). Early stopping helps prevent overfitting by halting the training when the model's performance on the validation set stops improving.

- Model Checkpoint

checkpoint = ModelCheckpoint(filepath=model_save_path, verbose=1, save_best_only=True, monitor='val_loss'): This callback saves the model after every epoch where there's an improvement in the monitored metric (in this case, val_loss). save_best_only=True ensures that only the model with the best performance on the validation set is saved. verbose=1 provides detailed logging about the checkpointing status. The model is saved to casting_product_detection.hdf5, as specified by model_save_path.

```python
model_save_path = 'casting_product_detection.hdf5'
early_stop = EarlyStopping(monitor='val_loss',patience=2)
checkpoint = ModelCheckpoint(filepath=model_save_path, verbose=1, save_best_only=True, monitor='val_loss')
```

```python
n_epochs = 20
results = model.fit_generator(train_set, epochs=n_epochs, validation_data=test_set, callbacks=[early_stop,checkpoint])
```

```python
model_history = { i:list(map(lambda x: float(x), j)) for i,j in results.history.items() }
with open('model_history.json', 'w') as f:
    json.dump(model_history, f, indent=4)
```

- Persistence: It allows for the training history to be persisted beyond the runtime of the current script, enabling later analysis or comparison with other models.
- Analysis: Having the training history in a human-readable file format like JSON makes it easier to analyze the model's performance across epochs, potentially using other tools or programming languages.

## Evaluate the Model

```python
pred_probability = model.predict_generator(test_set)
predictions = pred_probability > 0.5

plt.figure(figsize=(10,6))
plt.title("Confusion Matrix", size=20, weight='bold')
sns.heatmap(
    confusion_matrix(test_set.classes, predictions),
    annot=True,
    annot_kws={'size':14, 'weight':'bold'},
    fmt='d',
    xticklabels=['Defect', 'OK'],
    yticklabels=['Defect', 'OK'])
plt.tick_params(axis='both', labelsize=14)
plt.ylabel('Actual', size=14, weight='bold')
plt.xlabel('Predicted', size=14, weight='bold')
plt.show()
```

## Use the Model for prediciotn

```python
test_cases = ['ok_front/cast_ok_0_10.jpeg', 'ok_front/cast_ok_0_1026.jpeg', 'ok_front/cast_ok_0_1031.jpeg', 'ok_front/cast_ok_0_1121.jpeg', \
              'ok_front/cast_ok_0_1144.jpeg','def_front/cast_def_0_1059.jpeg', 'def_front/cast_def_0_108.jpeg', 'def_front/cast_def_0_1153.jpeg',\
              'def_front/cast_def_0_1238.jpeg', 'def_front/cast_def_0_1269.jpeg']

plt.figure(figsize=(20,8))
for i in range(len(test_cases)):
    img_pred = cv2.imread(test_path + test_cases[i], cv2.IMREAD_GRAYSCALE)
    img_pred = img_pred / 255 # rescale
    prediction = model.predict(img_pred.reshape(1, *image_shape))

    img = cv2.imread(test_path + test_cases[i])
    label = test_cases[i].split("_")[0]

    plt.subplot(2, 5, i+1)
    plt.title(f"{test_cases[i].split('/')[1]}\n Actual Label : {label}", weight='bold', size=12)
    # Predicted Class : defect
    if (prediction < 0.5):
        predicted_label = "def"
        prob = (1-prediction.sum()) * 100
    # Predicted Class : OK
    else:
        predicted_label = "ok"
        prob = prediction.sum() * 100

    cv2.putText(img=img, text=f"Predicted Label : {predicted_label}", org=(10, 30), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=0.8, color=(255, 0, 255), thickness=2)
    cv2.putText(img=img, text=f"Probability : {'{:.3f}'.format(prob)}%", org=(10, 280), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=0.7, color=(0, 255, 0), thickness=2)
    plt.imshow(img,cmap='gray')
    plt.axis('off')

plt.show()
```

## Misclassified Images
In machine learning, particularly in the field of computer vision, understanding the performance of a classification model extends beyond numerical metrics. One of the most insightful ways to evaluate model performance is to visually inspect the images that were misclassified. This approach allows us to identify patterns in the errors made by the model, offering clues on how to improve it.

```python
test_cases = ['ok_front/'+i for i in os.listdir(f'/content/drive/MyDrive/Colab Notebooks/Data/casting_data/test/ok_front')]
test_cases.extend(['def_front/'+i for i in os.listdir(f'/content/drive/MyDrive/Colab Notebooks/Data/casting_data/test/def_front')])

plt.figure(figsize=(20,8))
cnt = 0
misclassified = []
for i in range(len(test_cases)):
    if cnt == 10:break # show max 10 images
    img_pred = cv2.imread(test_path + test_cases[i], cv2.IMREAD_GRAYSCALE)
    try:
        img_pred = img_pred / 255 # rescale
    except:
        continue
    prediction = model.predict(img_pred.reshape(1, *image_shape))

    img = cv2.imread(test_path + test_cases[i])
    label = test_cases[i].split("_")[0]

    # Predicted Class : defect
    if (prediction < 0.5):
        predicted_label = "def"
        prob = (1-prediction.sum()) * 100
    # Predicted Class : OK
    else:
        predicted_label = "ok"
        prob = prediction.sum() * 100

    # select mismatched case
    if label != predicted_label:
        misclassified.append(test_cases[i])
        plt.subplot(2, 5, cnt+1)
        plt.title(f"{test_cases[i].split('/')[1]}\n Actual Label : {label}", weight='bold', size=12)
        cv2.putText(img=img, text=f"Predicted Label : {predicted_label}", org=(10, 30), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=0.8, color=(255, 0, 255), thickness=2)
        cv2.putText(img=img, text=f"Probability : {'{:.3f}'.format(prob)}%", org=(10, 280), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=0.7, color=(0, 255, 0), thickness=2)
        plt.imshow(img,cmap='gray')
        plt.axis('off')
        cnt += 1
    else:
        continue

plt.show()
```

## Explain Model
Interpreting SHAP (SHapley Additive exPlanations) visualizations, especially for deep learning models applied to images, involves understanding how individual pixels or regions of an image contribute to the model's output prediction. SHAP values quantify the impact of each feature (in this case, pixels) on the prediction relative to a baseline, allowing us to visually inspect which parts of an image lead the model to its conclusion.

Positive SHAP Values (Red): Areas of the image highlighted in red contribute positively towards the model's prediction. In a binary classification task, this means these areas push the model towards predicting one class over the other.

Negative SHAP Values (Blue): Conversely, areas highlighted in blue detract from that prediction, pushing the model towards the alternative class.
Intensity of the Color: The intensity or brightness of the color indicates the strength of the contribution. Brighter red or blue areas are more influential in the model's decision-making process than areas with faint colors.

```python
train_cases = ['ok_front/'+i for i in os.listdir('/content/drive/MyDrive/Colab Notebooks/Data/casting_data/train/ok_front')]
train_cases.extend(['def_front/'+i for i in os.listdir('/content/drive/MyDrive/Colab Notebooks/Data/casting_data/train/def_front')])
train_sample = [cv2.imread(train_path + i, cv2.IMREAD_GRAYSCALE).reshape(1, *image_shape) / 255 for i in np.random.choice(train_cases, 3000, replace=False)]
explainer = shap.DeepExplainer(model, train_sample[0])
```

```python
test_cases = ['def_front/cast_def_0_1059.jpeg', 'def_front/cast_def_0_108.jpeg', 'def_front/cast_def_0_1153.jpeg',\
              'def_front/cast_def_0_1238.jpeg', 'def_front/cast_def_0_1269.jpeg']

for i in test_cases:
    img = cv2.imread(test_path + i, cv2.IMREAD_GRAYSCALE).reshape(1, *image_shape) / 255
    shap_values = explainer.shap_values(img)
    shap.image_plot(shap_values, img, show=False)
    plt.title(f"{i.split('/')[1]}\n Actual Label : {i.split('_')[0]}", weight='bold', size=12)
    plt.axis('off')
plt.show()
```
